```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('raw.csv')
data
```

```python
n = 4 * 24 * 7 * 4
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(data['50Hertz'][:n], label='50Hertz')
ax.plot(data['Amprion'][:n], label='Amprion')
ax.plot(data['TenneT'][:n], label='TenneT')
ax.plot(data['TransnetBW'][:n], label='TransnetBW')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(data['Total'][:n], label='Total')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(data['Delta'][:n], label='Delta')
ax.legend()
```
